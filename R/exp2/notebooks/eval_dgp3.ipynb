```python
import os
import json
from glob import glob

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

## r"Absolute Error: | $CVR_{true}$ - $\bar{CVR_{1:t}}$ |"

```python
df_cvr = pd.read_csv(f'../{dpg_name}_noisy_true_cvr.csv')
```

```python
df_cvr.head()
```

```python
df_cvr = df_cvr.pivot(index='campaign_id', columns='channel', values='noisy_cvr').reset_index()

# Calculate avg CVR.
df_cvr['Facebook_AVG_CVR'] = df_cvr['Facebook'].cumsum() / df_cvr['campaign_id']
df_cvr['TV_AVG_CVR'] = df_cvr['TV'].cumsum() / df_cvr['campaign_id'
                                                     ]
df_cvr = df_cvr.rename(columns={'Facebook': '_Facebook', 'TV': '_TV'})
```

```python
sns.distplot(df_cvr['_Facebook'], label='Facebook')
sns.distplot(df_cvr['_TV'], label='TV')
plt.legend()
plt.xlabel('Noisy CVR',)
plt.grid()
```

```python
# Absolute Error.
df_cvr['Facebook'] = abs(df_cvr['Facebook_AVG_CVR'] - 0.05)
df_cvr['TV'] = abs(df_cvr['TV_AVG_CVR'] - 0.02)
```

```python
df_cvr.set_index('campaign_id')[['Facebook', 'TV']].plot(figsize=(12, 8), marker='.')
plt.legend(fontsize=18)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.ylabel(r"Absolute Error: | $CVR_{true}$ - $\bar{CVR_{1:t}}$ |", fontsize=20)
plt.xlabel('t', fontsize=20)
plt.grid()
```

## r"Mean Absolute Error: | $ROI_{true}$ - $\overline{ROI_{1:t}}$ |"

```python
# Set variables.
root_dir = f"/Users/s11616/Desktop/{dpg_name}"
```

```python
def ape(pred, true):
    return abs((pred - true) / true)
```

```python
def get_segmentname(row):
    cols = np.array(['Facebook', 'TV', 'Search'])
    flag = np.array([row[f'CALIBRATION_{x.upper()}'] for x in cols])
    if sum(flag) == 0:
        return '0. None'
    else:
        x = sum(np.arange(1, 4) * flag) + int(sum(flag) == 2) + int(sum(flag) == 3)
        return f'{x}. ' + '_'.join(cols[np.where(flag == 1)])
```

```python
rows = []
for exp_name in os.listdir(root_dir):
    jsonfile_paths = glob(os.path.join(root_dir, exp_name, '*.json'))
    
    for jsonfile_path in jsonfile_paths:
        
        jsonfile_name = jsonfile_path.split('/')[-1]
        if jsonfile_name == 'RobynModel-inputs.json':
            continue
        
        with open(jsonfile_path) as f:
            dic = json.load(f)
        
        for i, vals in enumerate(dic['ExportedModel']['summary']):
            if vals['variable'] in ['(Intercept)', 'trend', 'season', 'holiday'] :
                continue
            
            rows.append([exp_name, jsonfile_name, vals['variable'].split('_')[-1], vals['performance']])
```

```python
df = pd.DataFrame(rows)
df.columns = ['EXP_NAME', 'FILENAME', 'MEDIA', 'PRED_ROI']

df['DPG'] = dpg_name
```

```python
df_true = pd.read_csv(os.path.join(root_dir, os.listdir(root_dir)[0], 'true_roi.csv'))
_df_true = pd.DataFrame({'MEDIA': ['Facebook', 'TV'], 'TRUE_ROI': [float(x[0].split(': ')[1]) + 1 for x in df_true.values]})
```

```python
df = pd.merge(df, _df_true, on='MEDIA')
```

```python
df['APE_ROI'] = ape(df['PRED_ROI'], df['TRUE_ROI'])
```

```python
# Parse EXP_NAME.
df['USE_PROPHET'] = df['EXP_NAME'].apply(lambda x: x.split('_')[1] == "TRUE").astype(int)

df['CALIBRATION_FACEBOOK'] = df['EXP_NAME'].str.contains('Facebook').astype(int)
df['CALIBRATION_TV'] = df['EXP_NAME'].str.contains('TV').astype(int)
df['CALIBRATION_SEARCH'] = df['EXP_NAME'].str.contains('Search').astype(int)
df['T'] = df['EXP_NAME'].apply(lambda x: x.split('_')[-1]).astype(int)
```

```python
df.head()
```

```python
df_agg_models = df.groupby(['EXP_NAME', 'T', 'DPG', 'FILENAME']).mean()['APE_ROI'].reset_index()
```

```python
df_agg_models.head()
```

```python
plt.figure(figsize=(10, 8))
sns.boxplot(x = "T", y = "APE_ROI", data = df_agg_models.sort_values('T'), color=sns.color_palette()[0])
plt.title(f'{dpg_name}', fontsize=18)
plt.xticks(rotation=10)
plt.grid()
```

```python
plt.figure(figsize=(12, 6))
# sns.pointplot(x='T', y = 'APE_ROI', data = df_agg_models.sort_values('T'), errorbar="ci")
sns.lineplot(x='T', y = 'APE_ROI', data = df_agg_models.sort_values('T'), marker='o')
plt.ylim([0, 1.1])
plt.legend(fontsize=18)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.ylabel(r"Mean Absolute Error: | $ROI_{true}$ - $\overline{ROI_{1:t}}$ |", fontsize=16)
plt.xlabel('t', fontsize=16)
plt.grid()
```
